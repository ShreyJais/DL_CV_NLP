# 2. Data Preprocessing:

```python
import nltk
import re
from string import punctuation
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
def preprocess_text(text):
    regex_pattern = r"[a-z0-9\s]"
    result = re.findall(regex_pattern, text.lower())
    filtered_text = ''.join(result)
    tokens = nltk.word_tokenize(filtered_text)
    no_stop_words = [word for word in tokens if word.lower() not in stop_words]
    no_punctuation = [word for word in no_stop_words if word.lower() not in punctuation]
    return ' '.join(no_punctuation)
```

```python
df['image_description'] = df['image_description'].apply(preprocess_text)
```

tokenization

```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
max_length = 0
for index, description in enumerate(df.image_description):
  length = len(description)
  if length > max_length:
      max_length = length
print("Max length:", max_length)
```

```python
# Tokenization
max_length = 895
tokenized_texts = df['image_description'].apply((lambda x: tokenizer.encode(x, add_special_tokens=True, max_length=max_length, truncation=True)))
```

```python
from keras.preprocessing.sequence import pad_sequences
# Pad sequences
input_ids = pad_sequences(tokenized_texts, maxlen=max_length, dtype="int32", value=0, truncating="post", padding="post")
```

```python
attention_masks = np.where(input_ids != 0, 1, 0)
#attention masks to indicate which tokens are padding and which are real tokens. (here values!=0 are real tokens)
```

beart model

```python
from keras.models import Model
from keras.layers import Input
```

```python
for layer in bert_model.layers:
    layer.trainable = False
```

```python
input_ids = Input(shape=(max_length,), name='input_ids', dtype='int32')
#token_type_ids = Input(shape=(max_length,), dtype='int32')
input_masks = Input(shape=(max_length,), name='attention_masks', dtype='int32')
```

```python
embedding_layer = bert_model(input_ids, attention_mask=input_masks)[:1]
model = Model(inputs=[input_ids, input_masks], outputs=embedding_layer)
```
