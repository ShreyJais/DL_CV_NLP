# 2. Data Preprocessing:

```python
import nltk
import re
from string import punctuation
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
def preprocess_text(text):
    regex_pattern = r"[a-z0-9\s]"
    result = re.findall(regex_pattern, text.lower())
    filtered_text = ''.join(result)
    tokens = nltk.word_tokenize(filtered_text)
    no_stop_words = [word for word in tokens if word.lower() not in stop_words]
    no_punctuation = [word for word in no_stop_words if word.lower() not in punctuation]
    return ' '.join(no_punctuation)
```

```python
df['Name'] = df['name'].apply(preprocess_text)
```

vectorizing

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df[['name','size','brand','description']])
tfidf_matrix
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_df
```
